```python
from azureml.core import Workspace
from azureml.core.authentication import InteractiveLoginAuthentication
from azureml.core.datastore import Datastore
```

```python
ws = Workspace.from_config(path='azure_config.json', auth=InteractiveLoginAuthentication(tenant_id="5dbf1add-202a-4b8d-815b-bf0fb024e033"))
```

```python
datastore = Datastore.get_default(ws)
```

```python
import os
files_tr_blur = os.listdir('GoPro_v2/train/blur/data')
files_tr_blur = list(map(lambda x : os.path.join('GoPro_v2/train/blur/data/', x), files_tr_blur)) 
files_tr_sharp = os.listdir('GoPro_v2/train/sharp/data')
files_tr_sharp = list(map(lambda x : os.path.join('GoPro_v2/train/sharp/data/', x), files_tr_sharp)) 


files_te_blur = os.listdir('GoPro_v2/test/blur/data')
files_te_blur = list(map(lambda x : os.path.join('GoPro_v2/test/blur/data/', x), files_te_blur)) 
files_te_sharp = os.listdir('GoPro_v2/test/sharp/data')
files_te_sharp = list(map(lambda x : os.path.join('GoPro_v2/test/sharp/data/', x), files_te_sharp)) 

files = files_tr_blur + files_tr_sharp +files_te_blur + files_te_sharp
```
